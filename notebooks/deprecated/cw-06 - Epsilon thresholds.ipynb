```python
import pylab as plt
import numpy as np
import scipy.stats
```

```python
t = np.linspace(0, 5, 1000)
for s in np.arange(1, 21):
    T = scipy.stats.chi2.sf(t**2, 1)
    t2 = np.exp(-0.5*scipy.stats.chi2.isf(T, s))
    plt.semilogy(t, t2)
#plt.plot(t, T)
#plt.axhline(1e-6)
plt.grid()
plt.xlabel("Statistical significance, t [sigma]")
plt.ylabel("Threshold value, epsilon")
```
