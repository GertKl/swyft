```python
import scipy.stats
import numpy as np
import pylab as plt
```

```python
mu = np.linspace(-5, 5, 1000)
t = 5
sigma = 0.9
print((1-sigma**2)**0.5)
factor = scipy.stats.norm.cdf((t-mu)/sigma)-scipy.stats.norm.cdf((-t-mu)/sigma)
plt.plot(mu, factor)
plt.plot(mu, factor**10000)
```
